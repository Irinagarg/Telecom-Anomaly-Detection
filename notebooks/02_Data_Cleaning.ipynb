```python
# Cell 1 – Imports
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings("ignore")

# ── Must be in every notebook ──
def short_cell_name(co_dn):
    """
    Input : PLMN-PLMN/MRBTS-803406/LNBTS-803406/LNCEL-41
    Output: LNBTS-803406_LNCEL-41
    """
    parts = str(co_dn).split("/")
    if len(parts) >= 2:
        return f"{parts[-2]}_{parts[-1]}"
    return str(co_dn)

print("✅ Imports done")
```

```python
# Cell 2 – File path
file_path = r"../data/4G raw KPI Cell wise17th_June'26.xlsb"
CELL_COL  = "CO_DN"
```

```python
# Cell 3 – Load sheet names
excel = pd.ExcelFile(file_path, engine="pyxlsb")
print(f"Total sheets: {len(excel.sheet_names)}")
print(excel.sheet_names)   # ← was missing closing bracket
```

```python
# Cell 5 – Load all sheets into master DataFrame
df_list = []

for sheet in excel.sheet_names:
    print(f"  Reading: {sheet}")
    temp = pd.read_excel(
        file_path,
        sheet_name=sheet,
        engine="pyxlsb"
    )
    temp["Day"] = sheet
    df_list.append(temp)

print(f"\n✅ Loaded {len(df_list)} sheets")
```

```python
# Cell 6 – Concatenate all sheets
master_df = pd.concat(df_list, ignore_index=True)

# ── ADD THIS — create Short_Cell immediately after concat ──
master_df["Short_Cell"] = master_df[CELL_COL].apply(short_cell_name)

print(f"Master shape     : {master_df.shape}")
print(f"Unique CO_DN     : {master_df[CELL_COL].nunique()}")
print(f"Unique Short_Cell: {master_df['Short_Cell'].nunique()}")

# Confirm no collisions
if master_df[CELL_COL].nunique() == master_df["Short_Cell"].nunique():
    print("✅ Short_Cell is unique — no collisions")
else:
    diff = master_df[CELL_COL].nunique() - master_df["Short_Cell"].nunique()
    print(f"⚠️  {diff} collisions — check short_cell_name function")
```

```python
# Cell 7 – Shape check
print(f"Rows    : {master_df.shape[0]}")
print(f"Columns : {master_df.shape[1]}")
print(f"\nFirst 3 columns: {list(master_df.columns[:3])}")
print(f"Last  3 columns: {list(master_df.columns[-3:])}")
```

```python

# Cell 8 – Save master CSV
master_df.to_csv("../data/master_data.csv", index=False)
print(f"✅ Saved → ../data/master_data.csv")
print(f"   Shape: {master_df.shape}")
```

```python
# Cell 9 – Inspect column dtypes
print("Column dtype inspection:\n")
for col in master_df.columns[:20]:    # first 20 cols to keep output clean
    print(f"\n  {col}")
    print(f"  dtype  : {master_df[col].dtype}")
    print(f"  sample : {master_df[col].dropna().unique()[:5]}")
```

```python
# Cell 10 – Convert all columns to numeric where possible
# Skip identifier and day columns
skip_cols = [CELL_COL, "Short_Cell", "Day",
             "NE_STATE", "Short name"]

numeric_cols = [c for c in master_df.columns if c not in skip_cols]

for col in numeric_cols:
    master_df[col] = pd.to_numeric(master_df[col], errors="coerce")

print(f"✅ Converted {len(numeric_cols)} columns to numeric")
```

```python
# Cell 11 – Fill missing values with median per column
numeric_cols = master_df.select_dtypes(include=np.number).columns.tolist()
master_df[numeric_cols] = master_df[numeric_cols].fillna(
    master_df[numeric_cols].median()
)

missing_after = master_df[numeric_cols].isnull().sum().sum()
print(f"✅ Missing values after fill: {missing_after}")
```

```python
# Cell 14 – Cell-level aggregation (mean across all days)
# This is used in notebooks 04, 09 for ML modelling

kpi_cols = [
    "Average CQI_NBHD_Nokia_VeMr",
    "4g_Total_data_Traffic_4G_Radio",
    "E2E Call Setup Success Rate_Nokia_VeMr",
    "VoLTE Call Setup Success Rate_Nokia_VeMr",
    "Nokia_LTE_VoLTE Traffic Erl",
    "LPQUL.RSSI_PUSCH_AVG",
    "VoLTE Drop Call Rate_Nokia_VeMr"
]

# Only keep kpi_cols that actually exist
kpi_cols = [c for c in kpi_cols if c in master_df.columns]

# ── CHANGE: groupby Short_Cell instead of CO_DN ──
cell_df = (
    master_df.groupby("Short_Cell")[kpi_cols]
    .mean()
    .reset_index()
)

print(f"Cell-level aggregation shape: {cell_df.shape}")
print(f"Unique cells: {cell_df['Short_Cell'].nunique()}")
cell_df.head(5)
```

```python
# Cell 15 – Full cleaning pipeline
df = master_df.copy()

# Updated KPI list matching your actual column names from nb01 debug
kpi_cols = [
    "ERAB Drop Rate_Radio_Nokia_VeMr",
    "E2E Call Setup Success Rate_Nokia_VeMr",
    "VoLTE Call Setup Success Rate_Nokia_VeMr",
    "Average CQI_Nokia_VeMr",
    "Nokia_LTE_DL_User_Throughput_Mbps",
]
kpi_cols = [c for c in kpi_cols if c in df.columns]

for col in kpi_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df[kpi_cols] = df[kpi_cols].fillna(df[kpi_cols].median())

# ── Ensure Short_Cell exists ──
if "Short_Cell" not in df.columns:
    df["Short_Cell"] = df[CELL_COL].apply(short_cell_name)

print(f"Cleaned df shape : {df.shape}")
print(f"KPIs cleaned     : {kpi_cols}")
```

```python
# Cell 16 – Save cleaned data
# ── CHANGE: was missing Short_Cell in saved file ──
df.to_csv("../data/cleaned_master_data.csv", index=False)

# Also save as parquet for faster loading in later notebooks
df.to_parquet("../data/master_data.parquet", index=False)

print(f"✅ Saved → ../data/cleaned_master_data.csv")
print(f"✅ Saved → ../data/master_data.parquet")
print(f"   Shape     : {df.shape}")
print(f"   Columns   : {list(df.columns[:5])}...")
print(f"\nShort_Cell sample:")
print(df[[CELL_COL, "Short_Cell"]].drop_duplicates().head(5).to_string(index=False))
```
