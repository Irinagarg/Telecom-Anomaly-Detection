```python
# Cell 1 – Imports & Setup
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings("ignore")

# ── Unique cell identifier function ──
# Add this here so every subsequent cell in this notebook can use it
def short_cell_name(co_dn):
    """
    Creates unique cell identifier from CO_DN.
    Input : PLMN-PLMN/MRBTS-803406/LNBTS-803406/LNCEL-41
    Output: LNBTS-803406_LNCEL-41  (eNodeB_Cell — guaranteed unique)
    """
    parts = str(co_dn).split("/")
    if len(parts) >= 2:
        return f"{parts[-2]}_{parts[-1]}"
    return str(co_dn)

print("✅ Setup done")
```

```python
# Cell 4 – Load one sample sheet and inspect structure
sample_sheet = excel.sheet_names[0]
df_sample    = pd.read_excel(file_path, sheet_name=sample_sheet, engine="pyxlsb")

print(f"Sample sheet : {sample_sheet}")
print(f"Shape        : {df_sample.shape}  ({df_sample.shape[0]} cells × {df_sample.shape[1]} KPIs)")
print(f"\nColumn list:")
for i, col in enumerate(df_sample.columns):
    print(f"  {i:3d}  {col}")
```

```python
# Cell 5 – Check CO_DN and test short_cell_name
CELL_COL = "CO_DN"

if CELL_COL in df_sample.columns:
    print(f"Sample CO_DN values (first 5):\n")
    for val in df_sample[CELL_COL].dropna().head(5):
        print(f"  Full  : {val}")
        print(f"  Short : {short_cell_name(val)}")
        print()

    # Check uniqueness
    full_unique  = df_sample[CELL_COL].nunique()
    short_vals   = df_sample[CELL_COL].apply(short_cell_name)
    short_unique = short_vals.nunique()

    print(f"Unique cells (full CO_DN) : {full_unique}")
    print(f"Unique cells (short name) : {short_unique}")

    if full_unique == short_unique:
        print("✅ Short names are unique — no duplicates")
    else:
        print(f"⚠️  {full_unique - short_unique} collisions detected")
        print("   Check your CO_DN format — may need deeper split")
else:
    print(f"❌ CO_DN column not found")
    print(f"   Available columns: {list(df_sample.columns[:10])}")
```

```python
# Cell 7 – Load master_data.parquet (optimized)
parquet_path = "../data/master_data.parquet"

if os.path.exists(parquet_path):
    # Only load the 2 columns you need for this check
    # instead of loading all 80+ columns
    try:
        df_master = pd.read_parquet(
            parquet_path,
            columns=[CELL_COL]       # load only CO_DN first
        )

        df_master["Short_Cell"] = df_master[CELL_COL].apply(short_cell_name)

        print(f"Total rows           : {len(df_master)}")
        print(f"\nShort_Cell sample:")
        print(df_master[[CELL_COL, "Short_Cell"]].head(5).to_string(index=False))

        # Uniqueness check
        full_unique  = df_master[CELL_COL].nunique()
        short_unique = df_master["Short_Cell"].nunique()
        print(f"\nUnique (full CO_DN) : {full_unique}")
        print(f"Unique (Short_Cell) : {short_unique}")
        print("✅ No collisions" if full_unique == short_unique
              else f"⚠️ {full_unique - short_unique} collisions")

    except Exception as e:
        print(f"Could not load parquet: {e}")
else:
    print("master_data.parquet not found — will be created in notebook 02")
```

```python
import pandas as pd
import os

file_path = r"C:\Users\irina\Downloads\samplefile.xlsx"
output_path = r"C:\Users\irina\Downloads\samplefile_slim.xlsx"

xl = pd.ExcelFile(file_path)

KPI_COLS = [
    "CO_DN",
    "ERAB Drop Rate_Radio_Nokia_VeMr",
    "E2E Call Setup Success Rate_Nokia_VeMr",
    "VoLTE Call Setup Success Rate_Nokia_VeMr",
    "Average CQI_Nokia_VeMr",
    "Nokia_LTE_DL_User_Throughput_Mbps",
]

with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    for sheet in xl.sheet_names:
        df      = xl.parse(sheet)
        keep    = [c for c in KPI_COLS if c in df.columns]
        slim_df = df[keep].copy()
        slim_df.to_excel(writer, sheet_name=sheet, index=False)
        print(f"{sheet}: {len(df)} rows → kept {len(keep)} cols")

original = os.path.getsize(file_path)   / 1024 / 1024
slim     = os.path.getsize(output_path) / 1024 / 1024
print(f"\nOriginal : {original:.1f} MB")
print(f"Slim     : {slim:.1f} MB")
print(f"Reduced by: {((original-slim)/original*100):.0f}%")
print(f"\nSlim file saved → {output_path}")
```
